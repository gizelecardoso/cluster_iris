```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
sns.pairplot(df, hue="species")
```

```python
sns.pairplot(df)
```

```python
# Exluindo a coluna de species
# X -> caracteristicas de um objeto
X = np.array(df.drop('species', axis = 1))
```

# K-means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

## Comparando as classes criadas com as classes existentes de IRIS

```python
df['Kclasses'] = kmeans.labels_
```

```python
df
```

```python
sns.pairplot(df, hue='Kclasses')
```
